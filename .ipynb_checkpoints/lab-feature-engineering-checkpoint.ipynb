### Use appropriate methods to clean the columns GEOCODE2, WEALTH1, ADI, DMA,and MSA.

```python
# GEOCODE2
data["GEOCODE2"].isna().sum() 
```

```python
data["GEOCODE2"].value_counts()
```

```python
def clean_geocode2(x):
    x = str(x).lower()
    if x.startswith("a"):
        return "A"
    elif x.startswith ("b"):
        return "B"
    elif x.startswith ("c"):
        return "C"
    elif x.startswith ("d"):
        return "D"
    else: 
        return "A" #for unknown (NaN, empty) values I'll assign geocode A since it's the most common one.
```

```python
data["GEOCODE2"] = data["GEOCODE2"].apply(clean_geocode2)
data["GEOCODE2"].value_counts()
```

```python
# WEALTH1
data["WEALTH1"].value_counts()
```

```python
data["WEALTH1"].isna().sum() # The main problem with this column is a huge number of NaNs. Depending on the desired outcome we can replace is with 0 (assuming it wasn't provided because it was 0), or with median or mean. I'll use the mean nevertheless it is not a perfect situation considering the high number of missing values. By replacing them with the mean value we will underestimate the variance of this data.  
```

```python
data["WEALTH1"] = data["WEALTH1"].fillna(data["WEALTH1"].mean())
```

```python
# ADI
data["ADI"].isna().sum() #considering low number of missing values, I'll drop them. 
```

```python
data1 = data.dropna(subset=['ADI'])
```

```python
data1["ADI"].isna().sum()
```

```python
# DMA
data1["DMA"].isna().sum()
```

```python
data1["DMA"].value_counts()
```

```python
data1["DMA"] = data1["DMA"].astype(int) # changing from float to integer
```

```python
# MSA
data1["MSA"].isna().sum()
```

```python
data1["MSA"].value_counts()
```

```python
data1["MSA"] = data1["MSA"].astype(int) # changing from float to integer
```
